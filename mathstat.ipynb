```python
def statCrits(data): 
    return np.round(np.mean(data), 2), \
           np.round(np.std(data, ddof=1), 2), \
           np.round(np.var(data, ddof=1), 2), \
           np.round(len(data), 2)
```

```python
mean1, std1, var1, x1 = statCrits(denatr1)
```

```python
mean2, std2, var2, x2 = statCrits(denatr2)
```

```python
def corr(a,b):
    mean_a = np.mean(a)
    mean_b = np.mean(b)
    desp_a = 0
    desp_b = 0
    
    for i in a:
        desp_a = desp_a + (i - mean_a)
        
    for i in b:
        desp_b = desp_b + (i - mean_b)
        
    r = desp_a * desp_b / desp_a**2 / desp_b**2
    
    return r
        
```
